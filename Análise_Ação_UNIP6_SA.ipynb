```
plt.style.use('seaborn-darkgrid')
plt.figure( figsize=(16, 5) )
plt.title('Análise das ações da Unipar - Fechamento', fontsize=15, loc='left')
plt.plot( unipar_df.index, unipar_df['Close'] )

plt.xlabel('Período da Cotação')
plt.ylabel('Valor da Ação (R$)');
```

```
Media_Movel = unipar_df['Close'].rolling(5).mean()
Media_Tendencia = unipar_df['Close'].rolling(30).mean()

plt.style.use('seaborn-darkgrid')
plt.figure( figsize=(16, 5) )
plt.title('Análise das ações da Unipar - Fechamento', fontsize=15, loc='left')

plt.plot( unipar_df.index, unipar_df['Close'] )
plt.plot( unipar_df.index, Media_Movel )
plt.plot( unipar_df.index, Media_Tendencia )

plt.xlabel('Período da Coração')
plt.ylabel('Valor da Ação (R$)');
```
